```python
import pandas as pd


file_path = "C:/Users/lavan/OneDrive/Desktop/Activity1_customerChurn/preprocess.csv"
df = pd.read_csv(file_path)


df_info = df.info()
df_head = df.head()

df_info, df_head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


desc_stats = df.describe()


missing_values = df.isnull().sum()


corr_matrix = df.corr()


plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix")
plt.show()

desc_stats, missing_values
```

```python

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(df['tenure'], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Tenure')
sns.histplot(df['MonthlyCharges'], bins=30, kde=True, ax=axes[1], color='lightgreen')
axes[1].set_title('Distribution of MonthlyCharges')
sns.histplot(df['TotalCharges'], bins=30, kde=True, ax=axes[2], color='lightcoral')
axes[2].set_title('Distribution of TotalCharges')
plt.tight_layout()
plt.show()


fig, axes = plt.subplots(1, 3, figsize=(18, 5))


sns.boxplot(x='Churn', y='tenure', data=df, ax=axes[0], palette='coolwarm')
axes[0].set_title('Churn vs Tenure')


sns.boxplot(x='Churn', y='MonthlyCharges', data=df, ax=axes[1], palette='coolwarm')
axes[1].set_title('Churn vs MonthlyCharges')


sns.countplot(x='Contract', hue='Churn', data=df, ax=axes[2], palette='coolwarm')
axes[2].set_title('Churn by Contract Type')

plt.tight_layout()
plt.show()


fig, axes = plt.subplots(1, 3, figsize=(18, 5))


sns.countplot(x='gender', hue='Churn', data=df, ax=axes[0], palette='coolwarm')
axes[0].set_title('Churn by Gender')


sns.countplot(x='SeniorCitizen', hue='Churn', data=df, ax=axes[1], palette='coolwarm')
axes[1].set_title('Churn by SeniorCitizen')

sns.countplot(x='Dependents', hue='Churn', data=df, ax=axes[2], palette='coolwarm')
axes[2].set_title('Churn by Dependents')

plt.tight_layout()
plt.show()


fig, axes = plt.subplots(1, 2, figsize=(12, 5))


sns.countplot(x='InternetService', hue='Churn', data=df, ax=axes[0], palette='coolwarm')
axes[0].set_title('Churn by Internet Service')


sns.countplot(x='StreamingTV', hue='Churn', data=df, ax=axes[1], palette='coolwarm')
axes[1].set_title('Churn by StreamingTV')

plt.tight_layout()
plt.show()
```
